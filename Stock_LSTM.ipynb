##APPLE

```python
aapl = data1[['Open','High','Low','Close','Adj Close']]
```

```python
aapl
```

```python
dataset_train = aapl
```

```python
dataset_train = dataset_train.astype(float)
```

```python
# Using multiple features (predictors)
training_set_aapl = dataset_train.values
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
training_set_scaled_aapl = sc.fit_transform(training_set_aapl)

sc_predict = StandardScaler()
sc_predict.fit_transform(training_set_aapl[:,0:1])
```

```python
X_train = []
y_train = []

n_future = 60   # Number of days we want top predict into the future
n_past = 90  # Number of past days we want to use to predict the future

for i in range(n_past, len(training_set_scaled_aapl) - n_future +1):
    X_train.append(training_set_scaled_aapl[i - n_past:i, 0:training_set_aapl.shape[1] - 1])
    y_train.append(training_set_scaled_aapl[i + n_future - 1:i + n_future, 0])

X_train, y_train = np.array(X_train), np.array(y_train)

print('X_train shape == {}.'.format(X_train.shape))
print('y_train shape == {}.'.format(y_train.shape))
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
```

```python
# Initializing the Neural Network based on LSTM
regressor_aapl = Sequential()

# Adding 1st LSTM layer
regressor_aapl.add(LSTM(units=128, return_sequences=True, input_shape=(n_past,dataset_train.shape[1]-1)))

# Adding 2nd LSTM layer
regressor_aapl.add(LSTM(units=64, return_sequences=False))

# Adding 3rd layer
regressor_aapl.add(Dense(units=25))

# Output layer
regressor_aapl.add(Dense(units=1,activation='linear'))

# Compiling the Neural Network
regressor_aapl.compile(optimizer = Adam(learning_rate=0.01), loss='mean_squared_error')
```

```python
regressor_aapl.fit(X_train, y_train, shuffle = True,epochs = 100,validation_split=0.2, verbose=1, batch_size = 64)
```

```python
regressor_aapl.save('aapl_lstm.h5')
```
